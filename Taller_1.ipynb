```python
import pandas as pd
import os

if not os.path.exists("/content/drive/MyDrive/Diplomado/TALLER MODULO 1/resultados"):  # Punto 2
    os.makedirs("/content/drive/MyDrive/Diplomado/TALLER MODULO 1/resultados")
```

```python
datos = pd.read_excel("/content/drive/MyDrive/Diplomado/TALLER MODULO 1/DATOS/BaseEVA_Agrícola20192022.xlsx")
datos.head()
```

```python
cultivos = datos["Cultivo"].unique()   #punto 3
cultivos
```

```python
datos_meta = datos[datos["Departamento"] == "Meta"]   # punto 4
datos_meta. head()
```

```python
datos_meta = datos_meta[['Municipio', 'Cultivo', 'Año', 'Periodo', 'Área cosechada (ha)', 'Producción (t)', 'Rendimiento (t/ha)']]  # punto 5
datos_meta.head(2)
```

```python
datos_ordenados_area_cosechada = datos_meta.sort_values(by=["Área cosechada (ha)"], ascending = False)     # punto 6
datos_ordenados_area_cosechada.head()

datos_ordenados_area_cosechada.to_excel("/content/drive/MyDrive/Diplomado/TALLER MODULO 1/resultados/datos_ordenados_area_cosechada.xlsx", index = False)
```

```python
datos_meta.isnull().sum()   # punto 7
```

```python
datos_meta = datos_meta.fillna("Desconocido")   # punto 8
datos_meta.isnull().sum()

datos_meta.to_excel("/content/drive/MyDrive/Diplomado/TALLER MODULO 1/resultados/datos_meta_ajustado.xlsx", index = False)  # punto 9

```
